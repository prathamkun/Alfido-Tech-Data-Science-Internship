Feature Engineering

```python
photos['created dat'] = pd.to_datetime(photos["created dat"])
likes["created time"] = pd.to_datetime(likes["created time"])
comments["created Timestamp"] = pd.to_datetime(comments["created Timestamp"])
users["created time"] = pd.to_datetime(users["created time"])
```

```python
photos["Hour"] = photos["created dat"].dt.hour
photos["Day"] = photos['created dat'].dt.day_name()
photos["Month"] = photos["created dat"].dt.month_name()
```

```python
photos.head()
```

```python
likes_per_photo = (
    likes.groupby("photo").size().reset_index(name="Total Likes")

)
likes_per_photo.head()
```

```python
comments_per_photo = (
    comments.groupby("Photo id").size().reset_index(name="Total Comments")

)
comments_per_photo.head()
```

```python
engagement = photos.merge(
    likes_per_photo,
    left_on="id",
    right_on="photo",
    how="left"
)

engagement = engagement.merge(
    comments_per_photo,
    left_on="id",
    right_on="Photo id",
    how="left"
)
```

```python
engagement["Total Likes"] = engagement["Total Likes"].fillna(0)
engagement["Total Comments"] = engagement["Total Comments"].fillna(0)
```

```python
engagement["Total Engagement"] = (
    engagement["Total Likes"] +
    engagement["Total Comments"]
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

plt.hist(
    engagement["Total Likes"],
    bins=20,
    edgecolor="black"
)

plt.title("Distribution of likes")
plt.xlabel("Total likes")
plt.ylabel("Number of posts")

plt.show()
```

The histogram shows that most Instagram posts receive between 31 and 36 likes, with the highest concentration around 32–34 likes. Only a few posts have exceptionally high or low numbers of likes, indicating that engagement is relatively consistent across the dataset.

Since most posts receive a similar number of likes, Alfido Tech should focus on identifying the characteristics of the few high-performing posts (such as content type, posting time, or hashtags) to further improve engagement.

```python
plt.figure(figsize=(10,5))

plt.hist(
    engagement["Total Comments"],
    bins=20,
    edgecolor="black"
)

plt.title("Distribution of Comments")
plt.xlabel("Total Comments")
plt.ylabel("Number of Posts")

plt.show()
```

```python
hourly_engagement = (
    engagement.groupby("Hour")["Total Engagement"].mean().sort_index()

)
plt.figure(figsize=(12,5))

plt.plot(
    hourly_engagement.index,
    hourly_engagement.values,
    marker="o"
)

plt.title("Average Engagement by Posting Hour")
plt.xlabel("Posting Hour")
plt.ylabel("Average Total Engagement")

plt.xticks(range(24))
plt.grid(True)

plt.show()
```

All posts in the dataset were published at approximately 8 AM, resulting in only one posting hour being available for analysis. Because there is no variation in posting time, it is not possible to compare engagement across different hours.

Since every post was published around 8 AM, the dataset does not provide enough information to determine the optimal posting hour. Additional data containing posts published at different times would be required to identify the best posting schedule.

```python
hashtags = photo_tags.merge(
    tags,
    left_on="tag ID",
    right_on="id",
    how="left"
)

hashtags.head()
```

```python
# Count hashtag frequency

popular_tags = hashtags["tag text"].value_counts()

print(popular_tags.head(10))
```

```python
plt.figure(figsize=(10,5))

popular_tags.head(10).plot(
    kind="bar",
    edgecolor="black"
)

plt.title("Top 10 Most Popular Hashtags")
plt.xlabel("Hashtag")
plt.ylabel("Number of Uses")

plt.xticks(rotation=45)

plt.show()
```

The hashtag "smile" was the most frequently used hashtag in the dataset, appearing 59 times. Other commonly used hashtags include beach (42), party (39), and fun (38).

The remaining hashtags such as concert, food, lol, hair, happy, and beauty were used less frequently but still appeared consistently across posts.

This indicates that users tend to use lifestyle- and emotion-related hashtags more often than niche hashtags.

The analysis suggests that general lifestyle hashtags like smile, beach, party, and fun are widely used by Instagram users.

However, popularity alone does not necessarily mean these hashtags generate higher engagement. To make better marketing decisions, the next step is to analyze which hashtags are associated with the highest engagement rather than simply being used the most.

For Alfido Tech, hashtag selection should be based on both popularity and performance.

```python
engagement = engagement.drop(columns=["Photo id"])
```

```python
hashtag_engagement = hashtags.merge(
    engagement,
    on="photo",
    how="left"
)

hashtag_engagement.head()
```

```python
hashtag_performance = (
    hashtag_engagement
    .groupby("tag text")["Total Engagement"]
    .mean()
    .sort_values(ascending=False)
)

print(hashtag_performance.head(10))
```

```python
plt.figure(figsize=(10,5))

hashtag_performance.head(10).plot(
    kind="bar",
    edgecolor="black"
)

plt.title("Top 10 Highest Engagement Hashtags")
plt.xlabel("Hashtag")
plt.ylabel("Average Engagement")

plt.show()
```

The analysis shows that "delicious" has the highest average engagement (65.27), followed closely by "beauty" and "foodie". All top-performing hashtags have very similar engagement values, indicating that multiple hashtags perform consistently well rather than one hashtag dominating the dataset.

Businesses and content creators should select hashtags based on their engagement performance rather than only their popularity. Hashtags such as delicious, beauty, and foodie appear to attract slightly higher audience interaction and can help improve the visibility and engagement of future posts.
